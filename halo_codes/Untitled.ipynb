```python
# Load in the mass luminosity relation data

dat = np.loadtxt('/Users/jacobkennedy/Desktop/PHYS_459/data/Mh_ML_relations/L1600_vs_Mh_and_z.dat').T
Mh, L_1600_z6, L_1600_z7, L_1600_z8, L_1600_z9, L_1600_z10 = dat

# Given list of halo masses, would need to interpolate using np.interp
#halo_lumo_z7 = np.interp((halo_masses_z7), xp = Mh, fp = L_1600_z7)
#halo_lumo_z8 = np.interp((halo_masses_z8), xp = Mh, fp = L_1600_z8)
```

```python
fig, ax = plt.subplots(figsize=[14,8])
ax.loglog(Mh, L_1600_z6,'.', markersize=1, label = '$z = 6$')
ax.loglog(Mh, L_1600_z7,'.', markersize=1,label = '$z = 7$')
ax.loglog(Mh, L_1600_z8,'.', markersize=1,label = '$z = 8$')
ax.loglog(Mh, L_1600_z9,'.', markersize=1,label = '$z = 9$')
ax.loglog(Mh, L_1600_z10,'.', markersize=1,label = '$z = 10$')
ax.tick_params(axis='both', labelsize='20')
ax.set_xlabel(r'$\frac{M_h}{M_{Sun}}$', fontsize=35)
ax.set_ylabel(r'$L_{1600 \dot{A}} \, [\frac{erg}{s \cdot Hz}]$', fontsize=25)
ax.legend(fontsize=16)
fig.savefig('mass_lumo_halo.jpeg');
```
